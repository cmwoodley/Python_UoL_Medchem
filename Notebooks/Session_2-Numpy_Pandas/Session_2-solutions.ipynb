### Exercise 1 - reading in a pdb file

Read in a second pdb file - relative location is ./data/6atb.pdb

Iterate through the lines of this pdb and count the number of carbon atoms.

Print the percentage of atoms in the structure that are carbon atoms.

HINT: carbon in this instance can be matched by " C " - surrounded by spaces.

```python

# Read the PDB File
with open("./data/6atb.pdb", "r") as f:
    pdb_data = f.readlines()

# Count Carbon Atoms
atom_count = 0
carbon_count = 0
for line in pdb_data:
    if line.startswith("ATOM") :
        atom_count += 1
        if " C " in line:  # Ensure " C " is surrounded by spaces to match the element
            carbon_count += 1

# Output the Count
print("Total number of carbon atoms in 4ABT.pdb:", carbon_count)
print(f"Percentage of carbon atoms is: {100*carbon_count/atom_count:.2f}%")
```

### Exercise 4 - more numpy

Perform some simple mathematical functions on this toy dataset of atomic coordinates and charges

```python
# Sample data for atom coordinates and charges
# Coordinates are in Angstroms (A) and charges in elementary charge units (e)
atom_data = np.array([
    [1.0, 1.0, 1.0, 0.3],  # Atom 1: (x, y, z, charge)
    [1.5, 1.0, 1.5, -0.2], # Atom 2: (x, y, z, charge)
    [0.5, 2.0, 1.0, 0.1]   # Atom 3: (x, y, z, charge)
])

# 1. Print the average value of each column

mean_values = atom_data.mean(axis=0)

print("Mean position and charge: {}".format(mean_values))

#2. Print the average charge

mean_charge = np.mean(atom_data[:,3])

print("Mean charge: {:.3f}".format(mean_charge))

# 3. Calculate the distance between atom 1 and atom 2. Hint - euclidean distance = np.sqrt((x2-x1)**2+(y2-y1)**2+(z2-z1)**2)

distance = np.sqrt(np.sum((atom_data[1,:3] - atom_data[0,:3])**2))

print("Distance between atom 1 and 2: {:.3f}".format(distance))

# 4. print the maximum charge (np.max of a slice of the array) and the atom index with the highest **negative** charge.
#HINT see how we used argmax in the previous example. A similar function exists for the argmin (np.argmin)

max_charge = np.max(atom_data[:,3])
max_charge_id = np.argmin(atom_data[:,3])

print("Highest charge:", max_charge)
print("Atom with highest negative charge:", max_charge_id)

```

### Exercise 9 - Cleaning police data

The file ./data/police201501-11.csv contains information relating to driving offences in 2015 from months January to November.

Print the number of missing values per column and remove the columns containing **large numbers** of missing values.

Next remove rows containing missing values.

Import the data and clean it. Print the top 5 lines of the dataframe to the notebook with df.head()

```python
# Load in the csv
traffic_cops = pd.read_csv("./data/police201501-11.csv")

# Get the number of missing values by column
print(traffic_cops.isna().sum())

# Remove columns containig large numbers of missing values
traffic_cops = traffic_cops.drop(columns=["county_name","search_type"])
print("\nShape with columns removed:",traffic_cops.shape)

# Remove rows still contianing missing values
traffic_cops = traffic_cops.dropna()
print("\nShape with rows removed:",traffic_cops.shape)

# Ensure no missing values remain in the dataframe
print("\ntotal number of missing values:", traffic_cops.isna().sum().sum())
```

```python
# Now print the number of duplicates using df.duplicated.sum()
print("Number of duplicated entries:",traffic_cops.duplicated().sum())

# Now remove them
traffic_cops = traffic_cops.drop_duplicates()

traffic_cops.head()
```

### Exercise 10 - interrogating the data

Now we have a clean dataset we can do stuff with the data - not plotting because we'll save that for next week.

Answer the following questions using the data:

1. What unique outcomes were recorded?

2. What unique violations were recorded?

3. What is the mean age of a person stopped?

4. What was the most common violation?

5. Which gender was stopped most frequently?

6. Which race was most likely to get arrested if stopped?

```python
# What unique outcomes were recorded?

print("Unique outcomes:", np.unique(traffic_cops["stop_outcome"]))


# What unique violations were recorded?
print("\nUnique violations:", np.unique(traffic_cops["violation"]))

# What is the mean age of a person stopped? HINT: pandas dataframes have similar in-built functions to numpy arrays!
print("\nMean age:", traffic_cops["driver_age"].mean())
```

```python
# What is the most common violation?
unique_violations, violation_counts = np.unique(traffic_cops["violation"], return_counts=True)
most_common = np.argmax(violation_counts)
print(f"Most common violation is {unique_violations[most_common]} with {np.max(violation_counts)} counts.")

# Which gender was most commonly stopped?
unique_gender, gender_counts = np.unique(traffic_cops["driver_gender"], return_counts=True)
most_common = np.argmax(gender_counts)
print(f"Most common gender is {unique_gender[most_common]} with {np.max(gender_counts)} counts.")
```

```python
# Which race was most frequently arrested?

# Obtain unique values for race
races = np.unique(traffic_cops["driver_race"])

# Iterate through races with a loop, generate mask and calculate fraction of stops that led to an arrest
# hint - we can generate the dataframe slices similar to ar == 5 in the above and using df.loc[]

arrest_frequency = []
for race in races:
    indexes = traffic_cops["driver_race"] == race

    masked = traffic_cops.loc[indexes]

    # is_arrested is a boolean column, so when summing True is treated as 1 - this is how isna().sum() works
    num_arrests = masked["is_arrested"].sum()

    # Divide by total number of stops 
    total_stops = len(masked)

    arrest_frequency.append(num_arrests / total_stops)

# Put the results in a dataframe
    
freuency_by_race = pd.DataFrame({"driver_race":races,
                                 "Arrest_frequency":arrest_frequency})

freuency_by_race
```

### Exercise 11 - concatenating dataframes

We now have the data for December (./data/police201512.csv)

Load in the original dataset and the december dataset, concatenate them together using pd.concat(). Syntax is pd.concat([df1, df1], axis=0) where axis determines concat in rows or columns axis

Repeat your data cleaning steps.

Save the resulting dataset as ./data/police_2015.csv - syntax is df.to_csv("file_path.csv")

```python
# Load in the original csv
traffic_cops = pd.read_csv("./data/police201501-11.csv")
traffic_cops_dec = pd.read_csv("./data/police201512.csv")

traffic_cops_combined = pd.concat([traffic_cops, traffic_cops_dec], axis=0)

# Remove columns containig large numbers of missing values
traffic_cops_combined = traffic_cops_combined.drop(columns=["county_name","search_type"])

# Remove rows still contianing missing values
traffic_cops_combined = traffic_cops_combined.dropna()

# Now remove them
traffic_cops_combined = traffic_cops_combined.drop_duplicates()

traffic_cops_combined.tail()
```
